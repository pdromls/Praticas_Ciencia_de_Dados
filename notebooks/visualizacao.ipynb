```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as sts

ARQ = '../data/raw/co2.csv'

df2 = pd.read_csv(ARQ)

df2.head()
```

```python
unicos = list(set(df2.Treatment))
unicos
```

```python
# Percorre cada tipo de tratamento e plota um gráfico de dispersão
for tratamento in unicos:
    dados_tratamento = df2[df2.Treatment == tratamento]
    plt.scatter(dados_tratamento.conc, dados_tratamento.uptake, label=tratamento)
plt.legend(loc='lower right')
```
