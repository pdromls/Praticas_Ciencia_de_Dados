```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as sts

ARQ = '../data/raw/co2.csv'

df2 = pd.read_csv(ARQ)

df2.head()
```

```python
unicos = list(set(df2.Treatment))
unicos
```

```python
tipos = list(set(df2.Type))
tipos
```

```python
# Percorre cada tipo de tratamento e plota um gráfico de dispersão
for tratamento in unicos:
    dados_tratamento = df2[df2.Treatment == tratamento]
    plt.scatter(dados_tratamento.conc, dados_tratamento.uptake, label=tratamento)
plt.legend(loc='lower right')
```

```python
sns.scatterplot(x=df2.conc, y=df2.uptake, hue=df2.Type)
plt.show()
```

```python
q = df2.loc[df2.Type == 'Quebec']
m = df2.loc[df2.Type == 'Mississippi'] 

plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
sns.scatterplot(x=q.conc, y=q.uptake)
plt.title('Quebec')
plt.subplot(1,2,2)
sns.scatterplot(x=m.conc, y=m.uptake)
plt.title('Mississippi')
plt.tight_layout()
plt.show()
```

```python
ch = df2.loc[(df2.Treatment == 'chilled')]
nch = df2.loc[(df2.Treatment == 'nonchilled')]

plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
sns.scatterplot(x=ch.conc, y=ch.uptake)
plt.title('Chilled')
plt.subplot(1,2,2)
sns.scatterplot(x=nch.conc, y=nch.uptake)
plt.title('Nonchilled')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as sts

ARQ = '../data/raw/chicken.csv'

df4 = pd.read_csv(ARQ)

df4.head()
```
